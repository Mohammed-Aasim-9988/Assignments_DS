Apriori Algorithm

1) Association rules with 10% Support and 70% confidence

```python
# With 10% Support
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# with 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```



1.   A leverage value of 0 indicates independence.

2.   Range will be A high conviction value means that the consequent is highly depending on the antecedent and range [0 inf]



```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

DATA PREPROCESSING

```python
data1 = data.iloc[:,5:]
```

#Apriori Algorithm

1. Association rules with 10% Support and 70% confidence

```python
# with 10% support
frequent_itemsets=apriori(data1,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

#2. Association rules with 5% Support and 90% confidence

```python
# with 5% support
frequent_itemsets2=apriori(data1,min_support=0.05,use_colnames=True)
frequent_itemsets2
```

```python
# 90% confidence
rules2=association_rules(frequent_itemsets2,metric='lift',min_threshold=0.9)
rules2
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules2[rules2.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules2['support'],rules2['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```
