# K-MEANS

```python
from sklearn.cluster import KMeans
```

```python
from scipy.spatial.distance import cdist
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
data1 = pd.read_csv('crime_data.csv')
```

```python
#Normalized data fuction
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(data1.iloc[:,1:])
```

```python
# Elbow curv
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# selecting 4 clusters from above scree plot
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
data1['Clust']=x
```

```python
data1
```

```python
data1.iloc[:,1:5].groupby(data1.Clust).mean()
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
data3 = pd.read_csv("crime_data.csv")
```

```python
data3.info()
```

```python
data3.rename(columns={'Unnamed: 0': 'Place'}, inplace=True)
```

```python
data3
```

```python
data3.info()
```

```python
df = data3.iloc[:,1:5]
```

```python
df.values
```

```python
stdscaler=StandardScaler().fit(df.values)
x = stdscaler.transform(df.values)
```

```python
x
```

```python
dscan=DBSCAN(eps = 2,min_samples= 5)
dscan.fit(x)
```

```python
dscan.labels_
```

```python
cl = pd.DataFrame(dscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([data3,cl],axis = 1)
```
